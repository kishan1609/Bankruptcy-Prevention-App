```python
# Now we are reaedy for make a model
```

```python
x = new_df.drop('Class', axis='columns')
x.head()
```

```python
y = new_df.Class
y
```

```python
# Now, we are going to split dataset in train and test data.
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
# Now, we can go for prepare the model
```

```python
# 1. Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
model = lr.fit(x_train,y_train)
```

```python
model
```

```python
# Here we are going to predict the model
y_pred = model.predict(x_test)
y_pred
```

```python
# For chech the model Performence

from sklearn.metrics import confusion_matrix, classification_report
cm = confusion_matrix(y_test,y_pred)
cm
```

```python
print(classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
print("Accuracy:", accuracy_score(y_test,y_pred))
print("precision_score:", precision_score(y_test,y_pred))
print("Recall_score:", recall_score(y_test,y_pred))
print("F1-score:", f1_score(y_test,y_pred))
```

```python
model.score(x_test,y_test)
```

```python
model.score(x_train,y_train)
```

```python
# Based on accuracy and all other score, we can say that model is overfitted.
```
